## Import Data

```python
data = pd.read_csv(r"C:\Users\colit\Desktop\Master 1\Metode Avansate de Analiza a Datelor\Proiect - Random Forest\Dataset\winequality-red-fixed-csv.csv")
data.head(5)
```

## Split features and target (in our case quality of wine)

```python
X = data.drop('quality', axis=1)
Y = data['quality']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=41)
```

## Model

```python
rf = RandomForestClassifier()
rf.fit(X=X_train, y= Y_train)
```

```python
Y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
print(accuracy)
```
