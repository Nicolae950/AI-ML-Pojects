###### batch 8; learning rate 0.001

```python
model = YOLO('yolov8s.yaml')

results = model.train(data = os.path.join(ROOT_DIR, "1000_images_data_config.yaml"),
                      project = '/content/gdrive/MyDrive/Disertatie/Modele/yolov8s/sgd 50 epochs',
                      name = 'ep50_SGD_001lr_b8',
                      epochs = 50,
                      batch = 16,
                      imgsz = 640,
                      optimizer = 'SGD', # optimizator
                      seed = 42,
                      lr0 = 0.001, # initial learning rate
                      lrf = 0.001, # final learning rate
                      momentum = 0.835,
                      weight_decay = 0.00075,
                      val = True,
                      plots = True)
```

```python
# load trained model
trained_model = YOLO("/content/gdrive/MyDrive/Disertatie/Modele/yolov8s/ep50_SGD_001lr_b8/weights/best.pt")

metrics = trained_model.val(data = os.path.join(ROOT_DIR, "1000_images_data_config.yaml"),
                            batch = 8,
                            imgsz = 640,
                            conf = 0.40,
                            iou = 0.65,
                            plots = True)
```
